## Simple Returns - Part II

$$
\frac{P_1 - P_0}{P_0} = \frac{P_1}{P_0} - 1
$$

```python
import numpy as np
import yfinance as yf

# Download Microsoft stock data
MSFT = yf.download("MSFT", start="2015-01-01")

# Calculate simple returns
MSFT['simple_return'] = (MSFT['Close'] / MSFT['Close'].shift(1)) - 1
print(MSFT['simple_return'])
```

Plot the simple returns on a graph.

```python
import matplotlib.pyplot as plt
MSFT['simple_return'].plot(figsize=(9,6))
plt.show()
```

Calculate the average daily return.

```python
Average_daily_return = MSFT['simple_return'].mean()
Average_daily_return
```

Estimate the average annual return.

```python
Average_annual_return = Average_daily_return * 252
Average_annual_return
```

Print the percentage version of the result as a float with 2 digits after the decimal point.

```python
print (str(round(Average_annual_return, 4) * 100) + ' %')
```
